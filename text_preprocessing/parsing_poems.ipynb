```python
import requests
```

```python
URL = "https://www.gr-oborona.ru/pub/rock/egor_letov_stihi.html"
```

```python
response = requests.get(URL)
response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
html_poems = soup.find_all("pre")
len(html_poems)
```

```python
str_poems = [str(p) for p in html_poems]
str_poems[0]
```

```python
import re

str_poems = [re.sub(r'<[^>]+>', '', p.replace("\n"," ")) for p in str_poems]
```

```python
poems = [p.replace("    "," ").replace("  "," ")for p in str_poems]
poems
```

```python
with open('../data/text/poems.txt', 'w') as file:
    for item in poems:
        file.write(item + '\n')
```
